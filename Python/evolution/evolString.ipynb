# Evolution

This Jupyter notebook is a simple experiment in evolutionary programming with Python. We start with a simple conceptual test, how well randomized strings fit with a target string. First by a simple "count the number of correct letters in correct place"

```python

# Our target — what we're trying to evolve toward
TARGET = "sultans of swing"

# The alphabet our random strings can use
ALPHABET = string.ascii_lowercase + " "

def random_individual():
    """Create a random string the same length as TARGET."""
    return ''.join(random.choice(ALPHABET) for _ in range(len(TARGET)))

def fitness(individual):
    """Count how many characters match the target."""
    return sum(a == b for a, b in zip(individual, TARGET))

# Create a small population and score them
population = [random_individual() for _ in range(10)]

for p in population:
    print(f"{p}  |  fitness: {fitness(p)}/{len(TARGET)}")
```

And in this next cell, we use another approach. As a measure of fitness, we calculate how many positions apart the current letter is in relationship to the target letter. 


```python
# How far apart are two characters?
print(ord('h') - ord('g'))  # what do you get?
print(ord('h') - ord('z'))  # and this?
print(abs(ord('h') - ord('z')))  # why abs?
```

And after testing the concept test, lets make a function that calculates it, and apply that to a couple of test strings.

```python
test = "ulf"

guess = "str"

distances = [abs(ord(a) - ord(b)) for a,b in zip(test, guess)]

print(sum(distances))

def distanceFitness(input, target):
    distances = [abs(ord(a) - ord(b)) for a,b in zip(input, target)]
    return sum(distances)



print(distanceFitness("hello world", "hello world"))  # should be?
print(distanceFitness("gello world", "hello world"))  # one letter off by 1
print(distanceFitness("zzzzzzzzzzz", "hello world"))  # garbage
```

And now we create a number of random strings and compare them to the target string, and make a long list with the random strings and their fitness points. You want as low as possible. 

```python
population = [random_individual() for _ in range(10)]



scored = [(individual, distanceFitness(individual, TARGET)) for individual in population]

[print(a) for a in scored]

print("sorted")
sorted_scored = sorted(scored, key = lambda x: x[1])

[print(a) for a in sorted_scored]



```

## Change via heritage and mutation

We have two functions. One takes two parent strings, and randomly picks a cutoff point. A new string is created so that the first part comes from parent A and the second from parent B. And how much from each is decided by the cutoff point.

The second function takes a string and a percentage chance. Each letter have CHANCE probability to be exchanged with a random other letter. 

```python
def crossover(parent_a, parent_b):
    splitPoint = random.randint(0, len(parent_a)-1)
    result = parent_a[:splitPoint] + parent_b[splitPoint:]
    return result

def mutate(input, chance):
    output = ""
    for a in input:
        mutate_choice = random.randint(1, 100)
        if mutate_choice <= chance:
            output += random.choice(ALPHABET)
        else:
            output += a
    return output




```

# Final experiment

Here we create a large number of strings and store in Population. Then they are run through generations. For each generation, all the words are measured for how close they are to the target string. Then the bottom half is discarded, and the top half creates new strings by crossover and mutation (as above) These offspring are added to those who remained in this generation. Then a new generation starts, and all strings are ranked again. 

```python
population = [random_individual() for _ in range(100)]

for a in range(150):
    children = []
    mutated_children = []
    scored = [(individual, distanceFitness(individual, TARGET)) for individual in population]
    sorted_scored = sorted(scored, key = lambda x: x[1])
    kept = sorted_scored[:50]
    for i in range(0, len(kept)-1, 2):
        a = kept[i]
        b = kept[i+1]
        child = crossover(a[0], b[0])
        child2 = crossover(a[0], b[0])
        children.append(child)
        children.append(child2)
    for i in children:
        mutated_children.append(mutate(i,10))
    population = []
    population = [item[0] for item in kept] + mutated_children




[print(a)for a in population]





    

```
