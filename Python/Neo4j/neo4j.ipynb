```python
import json
import networkx as nx

# Load data
with open("nodes.json") as f:
    nodes = json.load(f)

with open("edges.json") as f:
    edges = json.load(f)

# Build graph
G = nx.DiGraph()

for node in nodes:
    node_id = node["id"]
    label = node.get("name", node_id)
    node_type = node["label"]
    G.add_node(node_id, label=label, node_type=node_type)

for edge in edges:
    G.add_edge(
        edge["from"], 
        edge["to"], 
        label=edge["type"],
        confidence=edge["confidence"]
    )

# Export
nx.write_gexf(G, "medieval.gexf")
print(f"Nodes: {G.number_of_nodes()}")
print(f"Edges: {G.number_of_edges()}")
print("Written to medieval.gexf")
```
