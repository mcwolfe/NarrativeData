# Base change example

```python
import numpy as np

A = np.array(
    [
        [66, 215],
        [95,38]
    ]
)

spiderman = np.array(
    [160,70]
)

coeffs = np.linalg.solve(A, spiderman)

# Print the result
print("Spider-Man in archetype basis:")
print(f"Doctor Strange component: {coeffs[0]:.3f}")
print(f"Hulk component: {coeffs[1]:.3f}")

# Optional: verify by reconstructing the original vector
reconstructed = A @ coeffs
print("\nReconstructed Spider-Man stats:", reconstructed)

```
