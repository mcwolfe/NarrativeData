```python
import networkx as nx

G = nx.Graph()
```

# Lets add rooms

```python
rooms = ["hallway", "kitchen", "living_room", "bedroom", "bathroom","study"]
for r in rooms:
    G.add_node(r, type="room")
```

# And doors

```python
doors = [
    ("hallway", "kitchen"),
    ("hallway", "living_room"),
    ("hallway", "bedroom"),
    ("hallway", "bathroom"),
    ("Living_room", "study")
]
G.add_edges_from(doors, relation="door")
```
