```python
from pint import UnitRegistry
import math
from sympy import symbols, pi, Rational, solve
import sympy.physics.mechanics as me
from sympy.physics import units

u = UnitRegistry()

# Define the wheel
radius = 3 * u.meter
rpm = 120 * u.rpm
density_iron = 7874 * u.kg / u.meter**3
thickness = 0.1 * u.meter  # let's assume 10cm thick


volume = math.pi * radius**2 * thickness
mass = (density_iron * volume).to(u.kg)
print(mass)


# Moment of inertia for solid disk: I = 0.5 * m * r²
I = (0.5 * mass * radius**2).to(u.kg * u.meter**2)
print(I)

import math

omega = (rpm).to(u.rad / u.second)
print(omega)

KE = (0.5 * I * omega**2).to(u.joule)
print(KE)

# Heating water from 20°C to 100°C
mass_water = 1 * u.kg
specific_heat = 4186 * u.joule / (u.kg * u.kelvin)
delta_T = 80 * u.kelvin

heat_energy = (mass_water * specific_heat * delta_T).to(u.joule)
print(f"Energy to heat water: {heat_energy}")

# Vaporization
latent_heat = 2.26e6 * u.joule / u.kg
vaporization_energy = (mass_water * latent_heat).to(u.joule)
print(f"Energy to vaporize: {vaporization_energy}")

total_energy = heat_energy + vaporization_energy
print(f"Total energy needed: {total_energy}")

print(f"Wheel energy / water energy: {(KE / total_energy).to(u.dimensionless)}")
```
