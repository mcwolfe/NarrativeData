```python
import simpy
import networkx as nx
import random


import networkx as nx
import random

sizes = [15, 20, 12, 18]
p = [
    [0.7,  0.05, 0.02, 0.15],
    [0.05, 0.6,  0.08, 0.03],
    [0.02, 0.08, 0.75, 0.10],
    [0.15, 0.03, 0.10, 0.5 ],
]

G = nx.stochastic_block_model(sizes, p, seed=42)
# Add travel times to our graph edges
for u, v in G.edges():
    G[u][v]['travel_time'] = random.randint(1, 5)

def walker(env, G, name, start_node):
    pos = start_node
    while env.now < 20:  # run until time 20
        neighbors = list(G.neighbors(pos))
        next_node = random.choice(neighbors)
        travel_time = G[pos][next_node]['travel_time']
        print(f"t={env.now:.1f} | {name} traveling from {pos} to {next_node} (takes {travel_time})")
        yield env.timeout(travel_time)
        pos = next_node
        print(f"t={env.now:.1f} | {name} arrived at {pos}")

env = simpy.Environment()
env.process(walker(env, G, "Walker A", 0))
env.process(walker(env, G, "Walker B", 1))
env.run()
```
